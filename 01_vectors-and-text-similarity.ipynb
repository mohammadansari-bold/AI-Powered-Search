````python
def cosine_similarity(vector_1, vector_2):
    """Calculate the cosine similarity between two vectors."""
    dot_product = np.dot(vector_1, vector_2)
    magnitude_vector_1 = np.linalg.norm(vector_1)
    magnitude_vector_2 = np.linalg.norm(vector_2)
    
    if magnitude_vector_1 == 0 or magnitude_vector_2 == 0:
        return 0.0
    
    cosine_similarity = dot_product / (magnitude_vector_1 * magnitude_vector_2)
    return cosine_similarity

def print_scores(scores):
    """Print the cosine similarity scores."""
    for i, score in enumerate(scores):
        print(f"Document {i + 1} similarity score: {score:.4f}")
````

````python
doc1_score = cosine_similarity(query_vector, doc1_vector)
doc2_score = cosine_similarity(query_vector, doc2_vector)
print_scores([doc1_score, doc2_score])
````

````python
def print_term_frequency_scores(terms, vectors, scores):
    """Print the term frequency scores for each document."""
    for i, doc in enumerate(vectors):
        print(f"doc{i + 1}: [" + ', '.join(map(lambda t : f"tf(doc{i + 1}, '{t}')", terms)) + "]")
    print("\Labels:", terms)

    for i, doc in enumerate(vectors):
        print(f" doc{i + 1}: [{', '.join(map(str, doc))}]")
    print("\Relevance Scores")

    for i, score in enumerate(scores):
        print(f"doc{i + 1}: {score}")
````

````python
def tf(term, doc):
    """Calculate the term frequency of a term in a document."""
    tokenized_doc = tokenize(doc)
    term_count = tokenized_doc.count(term.lower())
    document_length = len(tokenized_doc)
    return np.sqrt(term_count / document_length) if document_length > 0 else 0.0
````

````python
query = "cat in the hat"
terms = tokenize(query)
print(terms)
````

````python
query_vector = list(np.repeat(1, len(terms)))
print(query_vector)
````

````python
doc_vectors = [[tf(term, doc) for term in terms] for doc in docs]
doc_scores = [dot(query_vector, dv) for dv in doc_vectors]
````

````python
print_term_frequency_scores(terms, doc_vectors, doc_scores)
````

### To sum up till now

#### Document Frequency
------------------------
- Total number of document in the search engine that contain that the term
- It serves as measure of how important a term is
- intution here is rare words tend to be more important then common words like ```the``` and ```in```

$$DF(t \text{ in } d) = \sum_{d \text{ in } c} d.contains(t) ? 1 : 0$$

#### Inverse Document Frequency
--------------------------------
- Since we would like words which are more important to get a higher score

$$IDF(t \text{ in } d) = 1 + \log( totalDocs / ( DF(t) + 1 ) )$$


````python
def print_inverse_document_frequency_scores(terms, vectors):
    """IDF Vector values"""
    print(" [" + ', '.join(map(lambda t : f"idf('{t}')", terms)) + "]")
    print(f"IDF Vector:\n {vec2str(vectors)}")
````

````python
def idf(term):
    df_map = {"the": 9500, "cat": 100,
              "in":9000, "hat": 50}
    totalDocs = 10000

    return 1 + np.log(totalDocs / (df_map.get(term, 1)))

terms = ["the", "cat", "in", "the", "hat"]
idf_vector = [idf(term) for term in terms]

print_inverse_document_frequency_scores(terms, idf_vector)
````

### TF-IDF
- TF (measures how well a term describes a document)
- IDF (measures how important each term is)

$$TF\_IDF = TF * IDF^2$$

````python
def print_tf_idf_scores(terms, vectors, scores):
    print("Document TF-IDF Vector Calculations")
    for i, doc in enumerate(vectors):
        print(f" doc{i + 1}: [" + ', '.join(map(lambda t:
            f'tf_idf(doc{i + 1}, "{t}")',terms)) + "]")

    print("\nDocument TF-IDF Vector Scores")
    print ("Labels:", terms)
    for i, doc in enumerate(vectors):
        print(f"  doc{i + 1}: [{', '.join(map(num2str, doc))}]")

    print("\nRelevance Scores:")
    for i, score in enumerate(scores):
        print(f" doc{i + 1}: {num2str(score)}")
````

````python
def tf_idf(term, doc):
    return tf(term, doc) * idf(term)
````

````python
query = "The cat in the hat"
terms = tokenize(query)

query_vector = list(np.repeat(1, len(terms)))
doc_vectors = [[tf_idf(term, doc) for term in terms] for doc in docs]
doc_scores = [dot(query_vector, dv) for dv in doc_vectors]

print_tf_idf_scores(terms, doc_vectors, doc_scores)
````
